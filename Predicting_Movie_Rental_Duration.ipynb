**Data preprocessing**

```python
df['rental_date']=pd.to_datetime(df['rental_date'])
df['return_date']=pd.to_datetime(df['return_date'])
df.info()
```

```python
df['rental_length_days']=df['return_date']-df['rental_date']
df['rental_length_days']= df['rental_length_days'].dt.days
df.head()
```

```python
df['special_features'].value_counts()
```

```python
# Ensure special_features is of type str
df['special_features'] = df['special_features'].astype('str')

# Create the deleted_scenes column using apply with a lambda function
df['deleted_scenes'] = df['special_features'].apply(lambda x: 1 if 'Deleted Scenes' in x else 0)
df['behind_the_scenes'] = df['special_features'].apply(lambda x: 1 if 'Behind the Scenes' in x else 0)
# Print the DataFrame to verify the results
```

**Building the model**

```python
df.drop(columns=['rental_date','return_date','special_features'],inplace=True)
X=df.drop(columns=['rental_length_days'])
y=df['rental_length_days']
```

**Training and evaluating**

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=9)
best_model=LinearRegression()
best_model.fit(X_train,y_train)
y_pred=best_model.predict(X_test)
best_mse=MSE(y_test,y_pred)
print(best_mse)
```
